## Conversion to tflite
To use our model in an app - specifically flutter - we can't use the model as generated. That's why we convert it to tflite which can be used by many applications and frameworks. To do this we use a framework called ONNX. The ONNX format makes it possible to interact with our model as if it were a tensorflow lite model.
For further information: https://medium.com/geekculture/converting-yolo-v7-to-tensorflow-lite-for-mobile-deployment-ebc1103e8d1e

Install ONNX dependencies:

```python
!pip --quiet install onnx onnxruntime onnxsim
!pip install onnx-tf
```

Convert to ONNX

```python
!python yolov7/export.py --weights /content/runs/train/exp/weights/best.pt --grid --end2end --simplify \
        --topk-all 100 --iou-thres 0.65 --conf-thres 0.35 --img-size 640 640 --max-wh 640
```

```python
%cp /content/runs/train/exp/weights/best.onnx /content/drive/MyDrive
```

Convert to *tflite*

the -i defines the input path of our ONNX model and -o specifies the ouput directory

```python
!onnx-tf convert -i /content/runs/train/exp/weights/best.onnx -o /content/
```

```python
%cp /content/saved_model.pb /content/drive/MyDrive
```

Convert the tflite saved model to a .tflite file which we can use to make predictions

```python
import tensorflow as tf
```

```python
converter = tf.lite.TFLiteConverter.from_saved_model('/content/')
tflite_model = converter.convert()

with open('/content/yolov7_model.tflite', 'wb') as f:
  f.write(tflite_model)
```
